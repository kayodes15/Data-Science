### Of all the non numerical variables, the only correlated variable is Sex.

```python
subdf = df[['Pclass','Age','SibSp', 'Parch','Fare','Sex']]
subdf
```

### Sex dummy variable

```python
Sex_dummies = pd.get_dummies(subdf.Sex,prefix='Sex')
```

```python
df_with_dummies = pd.concat([subdf,Sex_dummies], axis='columns')
```

```python
df_with_dummies = pd.concat([subdf,Sex_dummies], axis='columns')
```

```python
df_with_dummies.drop(df_with_dummies[['Sex', 'Sex_male']], axis = 'columns', inplace = True)
df_with_dummies
```

```python
X = df_with_dummies
X
```

```python
y = df.Survived
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.3, random_state = 123)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
